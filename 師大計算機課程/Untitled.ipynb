```python
from math import *
from PIL import Image
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['simhei']
```

```python
original_img=Image.open("t0.jpg")
m,n=original_img.size[0],original_img.size[1]
print(f'圖片尺寸：寬為{m}，長為{n}')
plt.imshow(original_img)
plt.axis('off')
```

```python
%%timeit
new=Image.new('RGB',(m,n),(0,0,0))
def f1(x):
    return int(255*sqrt(x/255))

def f2(x):
    x=x/255  #轉換x 到0, 1之間
    y=3*x**(1/3)-3*x**(2/3)+x
    y=y**2
    return int(255*y)

def f(x):  #此x 必須在0, 1之間
    y=sin(x*pi/2)
    return y

for i in range(m):
    for j in range(n):
        r,g,b=im.getpixel((i,j))
        R=int(255*f(f(f(f(r/255)))))
        G=int(255*f(f(f(f(g/255)))))
        B=int(255*f(f(f(f(b/255)))))
        new.putpixel((i,j),(R,G,B))        
# new.save("test_C.jpg")
# print("OK")
plt.show()
```

```python
plt.figure(figsize=(8,10))
# plt.yticks([])
# plt.frame_on(False)
ax = plt.subplot(121)
ax.set_title('原圖')
plt.imshow(im)

ax2 = plt.subplot(122)
ax2.set_title('調整亮度')
plt.imshow(new)

for i in range(2):
    ax = plt.subplot(121+i)
    ax.set_xticks([])
    ax.set_yticks([])
    
plt.tight_layout() #調整空白


# 開啟/關閉外框
ax.set_frame_on(0) #True/False

plt.imshow(new)
```

### 針對暗部調亮

```python
from math import *
from PIL import Image
im=Image.open("t3.jpg")
m=im.size[0]
n=im.size[1]
new=Image.new('RGB',(m,n),(0,0,0))
c=[(0,0,0), (255,0,0),(255,255,0),(0,255,0),(0,255,255),
   (0,0,255),(255,0,255),(255,255,255)]

d=255*sqrt(3)/len(c)
for i in range(m):
    for j in range(n):
        r,g,b=im.getpixel((i,j))
        norm=sqrt(r**2+g**2+b**2)
        if 0<norm<2*d:
            x=norm/(2*d)
            nnorm=6*d*(x**0.5)
            R=nnorm*r/norm;G=nnorm*g/norm;B=nnorm*b/norm
        else:
            R=r;G=g;B=b

        new.putpixel((i,j),(int(R),int(G),int(B)))        
```

```python
plt.figure(figsize=(16,20))
# plt.yticks([])
# plt.frame_on(False)
ax = plt.subplot(121)
ax.set_title('原圖')
plt.imshow(im)

ax2 = plt.subplot(122)
ax2.set_title('調整亮度')
plt.imshow(new)

for i in range(2):
    ax = plt.subplot(121+i)
    ax.set_xticks([])
    ax.set_yticks([])
    
plt.tight_layout() #調整空白


# 開啟/關閉外框
ax.set_frame_on(0) #True/False

plt.imshow(new)
```

```python
from math import *
from PIL import Image
import numpy as np

im=Image.open("t4.jpg")
m=im.size[0]
n=im.size[1]
new=Image.new('RGB',(m,n),(0,0,0))
c=[(0,0,0), (255,0,0),(255,255,0),(0,255,0),(0,255,255),
   (0,0,255),(255,0,255),(255,255,255)]

d=255*sqrt(3)/len(c)
for i in range(m):
    for j in range(n):
        r,g,b=im.getpixel((i,j))
        norm=sqrt(r**2+g**2+b**2)-0.1
        R,G,B=c[int(norm/d)]
        new.putpixel((i,j),(R,G,B))        

```

```python
plt.figure(figsize=(16,20))
# plt.yticks([])
# plt.frame_on(False)
ax = plt.subplot(121)
ax.set_title('原圖')
plt.imshow(im)

ax2 = plt.subplot(122)
ax2.set_title('調整亮度')
plt.imshow(new)

for i in range(2):
    ax = plt.subplot(121+i)
    ax.set_xticks([])
    ax.set_yticks([])
    # 開啟/關閉外框
    ax.set_frame_on(0) #True/False
    
plt.tight_layout() #調整空白
plt.imshow(new)
```
